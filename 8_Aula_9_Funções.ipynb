```python
def taboada(numero):

  # Função que imprime a taboada para o número fornecido.

  print(f"Taboada do {numero}:")
  for i in range(1, 11):  # De 1 a 10
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")
```

Exercício 3

Você está desenvolvendo um sistema para uma loja que oferece um desconto progressivo com base no valor total das compras. A loja aplica descontos conforme a tabela abaixo:

- Para compras de até R$100,00: 5% de desconto.

- Para compras entre 100,01 e R$500,00: 10% de desconto.

- Para compras acima de R$500,00: 15% de desconto.

Crie uma função que receba o valor total da compra e retorne o valor final após aplicar o desconto correspondente. Além disso, a função deve retornar o valor do desconto aplicado.

image.png

```python
def desc_compras(compras):
  desconto = 0
  if compras <= 100:
    desconto = 0.05
  elif 100 < compras < 500:
    desconto = 0.1
  else:
    desconto = 0.15
  return desconto

for i in range(3):
  valor_1 = float(input("Digite o valor total da compra: R$ "))
  print(f"Desconto aplicado: R$ {desc_compras(valor_1)* valor_1}")
  print(f"Valor final após desconto: R$ {desc_compras(valor_1) * valor_1 + valor_1}\n")
```

Exercício 4

Você recebeu uma lista de dados de pilotos da Fórmula 1 com o número de corridas vencidas e a equipe de cada um em determinada temporada. Seu desafio é criar funções que analisem essas informações.

```python
pilotos = [
    {"nome": "Lewis Hamilton", "vitorias": 2, "equipe": "Mercedes"},
    {"nome": "Charles Leclerc", "vitorias": 3, "equipe": "Ferrari"},
    {"nome": "Lando Norris", "vitorias": 4, "equipe": "McLaren"},
    {"nome": "Sergio Perez", "vitorias": 4, "equipe": "Red Bull"},
    {"nome": "Carlos Sainz", "vitorias": 1, "equipe": "Ferrari"},
    {"nome": "George Russell", "vitorias": 1, "equipe": "Mercedes"},
    {"nome": "Esteban Ocon", "vitorias": 50, "equipe": "Alpine"},
    {"nome": "Fernando Alonso", "vitorias": 0, "equipe": "Aston Martin"},
    {"nome": "Pierre Gasly", "vitorias": 0, "equipe": "Alpine"},
    {"nome": "Max Verstappen", "vitorias": 15, "equipe": "Red Bull"},
]
```

Função 1: pilotos_com_vitorias(pilotos, min_vitorias)

Retorna uma lista com o nome dos pilotos que possuem mais ou igual a min_vitorias

```python
def pilotos_com_vitorias(pilotos, min_vitorias):
  lista_min = []
  for chave in pilotos:
    if chave['vitorias'] >= min_vitorias:
      lista_min.append(chave['nome'])

  return lista_min

print(pilotos_com_vitorias(pilotos, 3))
```

Função 2: piloto_mais_vitorias(pilotos)

Retorna o nome do piloto com mais vitórias.

```python
def piloto_mais_vitorias(pilotos):
  lista_max = []
  indice = []

  for chave in pilotos:
      lista_max.append(chave['vitorias']) # adiciona o número de vitórias de cada piloto em uma lista para ser usado o max() depois
      indice.append(chave['nome']) # salva o nome de todos o pilotos com o mesmo índice do número de vitórias
      cursor = lista_max.index(max(lista_max)) # encontra o maior nº de vitórias e retorna o indice

  return indice[cursor] # Usa o mesmo indice encontrado no cursor e aplica na lista de nomes

piloto_mais_vitorias(pilotos)
```
