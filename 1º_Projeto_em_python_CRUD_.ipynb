# Software de gerenciamento de equipamentos


Decidi fazer um código um pouco diferente da atividade, porém eu ainda vou usar todos os requisitos listados nela. Como eu trabalho com manutenção, eu vou fazer um gerenciador de equipamentos simples, e cada equipamento deve ter:

Dicionário do json:

- Ordem de serviço: int

- Modelo de equipamento: string

- Tensão: int + "V" - ex: 220V, 380V, 440V.

- Corrente: int + "A"

Pelo o que eu pesquisei, esse "tipo" de programa é geralmente chamado de **CRUD** (*Create, Read, Upload e Delete, são as 4 funções principais desses softwares*)


Aqui estão os requisitos originais da atividade:


- Adicionar uma nova tarefa (Novo equipamento)

- Visualizar todas as tarefas (Apenas OS e Modelo)

- Remover uma tarefa existente (Remover Equipamento por OS)

- Sair do programa


As tarefas devem ser armazenadas em um arquivo de texto (usarei .json), de forma que, ao reabrir o programa, as tarefas previamente salvas sejam carregadas automaticamente. Cada tarefa pode ser representada por uma simples linha de texto (no meu caso, usarei um dict para cada equipamento).


Requisitos:

- O programa deve exibir um menu interativo com as opções.

- Deve permitir a inclusão de tarefas (equipamentos) por meio de entrada do usuário.

- As tarefas devem ser exibidas com numeração (usarei o nº de OS) quando o usuário optar por visualizá-las.

- O usuário deve conseguir remover uma tarefa selecionando o número correspondente.

- As tarefas devem ser salvas em um arquivo (equipamentos.json) e carregadas sempre que o programa for iniciado.

- Organize o programa em funções para cada funcionalidade.

OBS: Eu usei a IA do chatGPT em alguns pontos, como o encoding 'utf8' e alguns erros que não consegui achar uma solução, deixei anotado onde usei e onde foi escrito normalmente


# Código e python

```python
import json

def salvar_json_format(nome_arquivo, lista_dados):
  # Esta função cria um arquivo .json

  with open(nome_arquivo, 'w', encoding='utf-8') as file:
    json.dump(lista_dados, file, indent=2) # O indent serve para adicionar recuos no arquivo json

def adicionar_linha(nome_arquivo, nova_linha):
# Esta função adiciona o dicionário criado com o criar_dict() ao arquivo .json

  with open(nome_arquivo, 'r', encoding='utf-8') as file:
    lista_var = json.load(file)

  lista_var.append(nova_linha)

  with open(nome_arquivo, 'w', encoding='utf-8') as file:
    json.dump(lista_var, file, indent=2)

def criar_dict(data_1, data_2, data_3, data_4):

# Cria os dicionários antes de adicionar a lista que vai para o json

# Antes eu tinha feito a função assim:

# def criar_dict(data_1, data_2, data_3, data_4):
#   dicionario = {}
#   dicionario['OS'] = data_1
#   dicionario['modelo'] = data_2
#   dicionario['tensao'] = f"{data_3}V"
#   dicionario["corrente"] =f"{data_4}A"

# Mas com ajuda da IA, descobri que o return pode formatar o dict sem declarar ele antes:

  return {
    "OS": data_1,
    "modelo": data_2,
    "tensao": f"{data_3}V",
    "corrente": f"{data_4}A"
}

def remover_cadastro_nome(nome_arquivo, chave):
  with open(nome_arquivo, 'r', encoding='utf-8') as file:
    lista_var = json.load(file)

  indice = None

  for i, cadastro in enumerate(lista_var): # O enumerate já adiciona um indice ao dicionário dentro da lista do json
    if cadastro['OS'] == chave:
        indice = i
        break

  if indice != None:

    lista_var.pop(indice) # Apaga o item da lista de indice

    with open(nome_arquivo, 'w', encoding='utf-8') as file:
        json.dump(lista_var, file, indent=2)# Reescreve toda a lista, mas agora sem o item que foi apagado
  else:
    print("Erro - nome não encontrado")

def cadastro_equip():

  os = int(input("Digite o número de Ordem de Serviço: "))
  model = input("Digite o modelo do equipamento: ")
  volt = int(input("Digite a tensão nominal do equipamento: "))
  amp = int(input("Digite a corrente nominal do equipamento: "))

  return os, model, volt, amp # Mesmo sem declarar uma tupla, isso retorna uma tupla(os, model, volt, amp)

def interface():
  try:
    selection = int(input('''
------Gerenciador de equipamento------

Digite para selecionar uma das opções:

1 - Adicionar novo equipamento
2 - Visualizar todos equipamentos (OS e modelo)
3 - Remover equipamento
4 - Pesquisar ordem de serviço
0 - sair
'''))
  except ValueError:
    return 99 # o 99 é só para levantar uma falha caso haja algum erro e ser tratado em uma condicional lá no main()
  return selection

def exibir(nome_arquivo):
  with open(nome_arquivo, 'r', encoding='utf-8') as file:
    lista_var = json.load(file)

    ordenada = sorted(lista_var, key=lambda i: i["OS"]) #Ordena a lista usando as chaves de 'OS' como valor de ordenação

  for chave in ordenada:
    print(f"Ordem de serviço: {chave['OS']}")
    print(f"Modelo: {chave['modelo']}")
    print("\n")

def pesquisa(nome_arquivo, chave):
  with open(nome_arquivo, 'r', encoding='utf-8') as file:
    lista_var = json.load(file)

  encontrado = 0

  for cadastro in lista_var:
    if cadastro['OS'] == chave:
      print(f"Ordem de serviço: {cadastro['OS']}")
      print(f"Modelo: {cadastro['modelo']}")
      print(f"Tensão nominal: {cadastro['tensao']}")
      print(f"Corrente nominal: {cadastro['corrente']}\n")
      encontrado = 1
      break

  if encontrado == 0:
    print("Equipamento não encontrado!")

def main():

  nome = "equipamento.json"  # corrigi aqui para ser usado como caminho do arquivo
  salvar_json_format(nome, [])  # cria JSON vazio

  while True:
    try:
      sel = interface()
      if sel == 1:
        tupla_c = cadastro_equip()
        dict_c = criar_dict(*tupla_c) #Aqui o chatGPT ajudou, antes era assim (tupla_c[0], tupla_c[1]...)
        adicionar_linha(nome, dict_c)
        print('Item adicionado!\n')

      elif sel == 2:
        print("\n")
        exibir(nome)

      elif sel == 3:
        apagar = int(input("Digite o número da OS que deseja remover: "))
        remover_cadastro_nome(nome, apagar)

      elif sel == 4:
        pesquisa_1 = int(input("Digite o número de OS: "))
        pesquisa(nome, pesquisa_1)

      elif sel == 0:
        break

      elif sel == 99: # O 99 é retornado pelo except ValueError lá de cima, na def interface()
        print("Entrada inválida, tente novamente.")

    except Exception as e: #Igual na aula 4 com o uso do nome da Exception como variável
      print(f"Erro: {e}")

main()
```
